```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df
```

```python
df.isnull().sum()
```

```python
df.describe()
df.info()
```

```python
df.dropna()
```

```python
median=df['math score'].median()
print(median)
df['math score']=df['math score'].fillna(median)
df
```

```python
df.boxplot()
```

```python
newdf=df[df['math score'] > 30]
newdf
```

```python
newdf.boxplot()
```

```python
sns.histplot(df['math score'], kde=True)
plt.title('Math Score Distribution (Before Transformation)')
plt.show()
print("Skewness Before Transformation:", df['math score'].skew())
```

```python
df['math score log'] = np.log1p(df['math score'])  # log(1 + x) to handle zero values safely
```

```python
sns.histplot(df['math score log'], kde=True)
plt.title('Math Score Distribution (After Log Transformation)')
plt.show()
print("Skewness After Transformation:", df['math score log'].skew())
```
